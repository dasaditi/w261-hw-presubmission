__What is Apache Spark and how is it different to Apache Hadoop? __   
Spark is a fast and general-purpose cluster computing system, which means that it competes more with MapReduce than with the entire Hadoop ecosystem. For example, Spark doesn't have its own distributed filesystem, but can use HDFS.

## Algorithm Weighted k-means
__Input__: a set of n data points obtained from the density-biased reservoir sampling, and the number of clusters (K)    
__Output__: centroids of the K clusters    

1) Initialize the K cluster centers   
2) Repeat: Assign each data point to its nearest cluster center according to the membership function,

$$
m(c_j|x_i) = \frac{|| x_i - c_j ||^{-p-2}}{\sum_{j=1:k} || x_i - c_j||^{-p-2}}
$$

3) For each center $c_j$, recompute the cluster center c_j using the current cluster memberships and weights,

$$
c_j = \frac{\sum_{j=1:n} m(c_j|x_i) w( x_i) x_i }{ \sum_{j=1:n} m(c_j|x_i) w( x_i)}
$$

where $w(x_i)$ is a weight associated with each data point   
    4) Until there is no reassignment of data points to new cluster centers
    
Ref: http://science.sut.ac.th/mathematics/pairote/uploadfiles/weightedkm-temp2_EB.pdf

````python
############################################################################################################
# http://nbviewer.jupyter.org/urls/dl.dropbox.com/s/3nsthvp8g2rrrdh/EM-Kmeans.ipynb
############################################################################################################
from __future__ import division
import numpy as np
import time
import copy
import numpy as np

#Calculate which class each data point belongs to
def nearest_centroid(line):
    x = np.array([float(f) for f in line.split(',')])
    closest_centroid_idx = np.sum((x - centroids_Broadcast.value)**2, axis=1).argmin()
    w = np.sqrt(np.sum(x**2))
    return (closest_centroid_idx,(x/w,1/w))


K = 3
# Initialization: initialization of parameter is fixed to show an example
centroids = np.array([[0.0,0.0],[2.0,2.0],[0.0,7.0]])

D = sc.textFile("./data.csv").cache()
iter_num = 0
start = time.time()
for i in range(10):  
    centroids_Broadcast = sc.broadcast(centroids) 
    res = D.map(nearest_centroid).reduceByKey(lambda x,y : (x[0]+y[0],x[1]+y[1])).collect()
    res = sorted(res,key = lambda x : x[0])  #sort based on clusted ID
    centroids_new = np.array([x[1][0]/x[1][1] for x in res])  # divide by cluster size
    
    if np.sum(np.absolute(centroids_new-centroids_Broadcast.value)) < 0.01:
        print "done after iteration " + str(iter_num+1)
        print centroids
        plot_iteration(centroids)
        break
    
    
    iter_num = iter_num + 1 
    centroids = centroids_new
    print "Iteration " + str(iter_num)
    print centroids
    plot_iteration(centroids)

end = time.time()
print "Time to converge: ", end-start, "seconds"
print "Final Results:"
print centroids
````

### Linear Regresssion
#### Mean Squared Error loss
The square loss function is both convex and smooth and matches the 0–1 indicator function when $yf(\vec  x)=0$ and when $yf({\vec  {x}})=1$. However, the square loss function tends to penalize outliers excessively, leading to slower convergence rates (with regards to sample complexity) than for the logistic loss or hinge loss functions. In addition, functions which yield high values of $f({\vec {x}})$ for some $x\in X$ will perform poorly with the square loss function, since high values of $yf({\vec  {x}})$ will be penalized severely, regardless of whether the signs of $y$ and $f({\vec {x}})$ match.

#### Objective Function with ridge regularization term

\begin{equation}
arg min_w~L(w) = \sum_{i=1}^n\left(y_i-w^T x_i\right)^2 + \lambda||\vec w||^2
\end{equation}

#### Gradient Descent
\begin{equation}
\frac{\partial L}{\partial w} = \frac{1}{n} \sum_{i = 0}^{n-1} (w^T x_i - y_i) \cdot x_i + \lambda w
\end{equation}

RSS:

$$RSS = \sum_{i=1}^n\left(y_i - \beta_0 - \sum_{j=1}^p\beta_jx_{ij}\right)$$

Ridge regularization term:
$$L2 = \lambda\sum_{i=1}\beta_i^2$$

Lasso regularization term:
$$L1 = \lambda\sum_{i=1}|\beta_i|$$

Ref: http://www-bcf.usc.edu/~gareth/ISL/ISLR%20Fourth%20Printing.pdf CHAPTER 6.2

````python
import numpy as np
def linearRegressionGD(data, wInitial=None, learningRate=0.05, iterations=50, regParam=0.01, regType="Lasso"):
    featureLen = len(data.take(1)[0])-1
    n = data.count()
    if wInitial is None:  #start learning from a random vector
        w = np.random.normal(size=featureLen) # w should be broadcasted if it is large
    else:                 #start from provided vector
        w = wInitial
    for i in range(iterations):
        wBroadcast = sc.broadcast(w)   #make available in memory as read-only to the executors (for mappers and reducers)
        gradient = data.map(lambda d: -2 * (d[0] - np.dot(wBroadcast.value, d[1:])) * np.array(d[1:])) \
                    .reduce(lambda a, b: a + b)
            
            
        if regType == "Ridge":
            wReg = wBroadcast.value * 1
        elif regType == "Lasso":
            wReg = wBroadcast.value * 1
            wReg = (wReg>0).astype(int) * 2-1 ## is 0/1; doubling is 0/2; subtracting 1 is -1/1   
        else:
            wReg = np.zeros(wBroadcast.value.shape[0])   
            
        gradient = gradient + regParam * wReg  #gradient:  GD of Sqaured Error+ GD of regularized term   
        
        w = w - learningRate * gradient/n
        with open('linearRegressionGD_weights_'+str(regType)+'.txt', 'a') as f:
            f.write(str(w[0])+","+str(w[1])+"\n")
    return w
#http://www.analyticsvidhya.com/blog/2016/01/complete-tutorial-ridge-lasso-regression-python/
````
