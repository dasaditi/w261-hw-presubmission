### CACHE
Data can be cached within each executor (worker), such that if the worker needs to iterate over the data more than once, it does not have to read from disk again. Cached variables are STATIC.


### BROADCAST
Broadcast variables are book keeping information for the learning algorithm that we wish to share with each executor, like a weight vector. Broadcast variables are DYNAMIC - they are updated at each iteration. Broadcasting can have a significant positive impact on performance when there are many more workers than mappers, as when the dataset is large. 

#### Example:
900Gig of data: assume 1Gig chunks   
– Ship (serialize) each mapper and centroids data to each executor 900 times

````python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt


def modified_huber_loss(y_true, y_pred):
    z = y_pred * y_true
    loss = -4 * z
    loss[z >= -1] = (1 - z[z >= -1]) ** 2
    loss[z >= 1.] = 0
    return loss

plt.figure(figsize=(10,7))
xmin, xmax = -4, 4
xx = np.linspace(xmin, xmax, 100)
plt.plot([xmin, 0, 0, xmax], [1, 1, 0, 0], 'k-',
         label="Zero-one loss")
plt.plot(xx, np.where(xx < 1, 1 - xx, 0), 'g-',
         label="Hinge loss")
plt.plot(xx, -np.minimum(xx, 0), 'm-',
         label="Perceptron loss")
plt.plot(xx, np.log2(1 + np.exp(-xx)), 'r-',
         label="Log loss")
plt.plot(xx, np.where(xx < 1, 1 - xx, 0) ** 2, 'b-',
         label="Squared hinge loss")
plt.plot(xx, modified_huber_loss(xx, 1), 'y--',
         label="Modified Huber loss")
plt.ylim((0, 8))
plt.legend(loc="upper right")
plt.xlabel(r"Decision function $f(x)$")
plt.ylabel("$L(y, f(x))$")
plt.show()
````

````python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import numpy as np
import matplotlib.pyplot as plt

# draw the learned surface for linear SVM and logistic regression
def sep_surface():  
    
    # reuse the code from 11.3 to create linearly separable dataset
    xb = (np.random.rand(50)*2-1)/2-0.5
    yb = (np.random.rand(50)*2-1)/2+0.5
    xr = (np.random.rand(50)*2-1)/2+0.5
    yr = (np.random.rand(50)*2-1)/2-0.5
    inputs = []
    yy = []
    for i in range(len(xb)):
        inputs.append([xb[i],yb[i], 1])
        inputs.append([xr[i],yr[i], -1])
    
    # set up axes for weight space
    w0 = np.linspace(-10, 10, 100)
    w1 = np.linspace(-10, 10, 100)
    
    # set up 2-d arrays to store values for z axis
    # each entry is sum of loss function across dataset for specific value of w0 and w1
    hinge_z = np.zeros((100, 100))
    log_z = np.zeros((100, 100))
    for i in range(w0.shape[0]):
        for j in range(w1.shape[0]):
            for k in inputs:
                # calculate hinge loss first for linear SVM
                hinge = 1 - k[2]*(w0[i]*k[0] + w0[j]*k[1])
                if hinge < 0:
                    hinge = 0
                hinge_z[i][j] += hinge
                
                # do the same for log loss
                #log_loss = np.log(1 + np.exp(-k[2]*(w0[i]*k[0] + w0[j]*k[1])))
                log_z[i][j] += np.log(1 + np.exp(-k[2]*(w0[i]*k[0] + w0[j]*k[1])))
    
    # plot the surfaces in separate figures
    w0, w1 = np.meshgrid(w0, w1)
    fig = plt.figure(figsize=(14, 7))
    ax = fig.add_subplot(1, 2, 1, projection='3d')
    surf = ax.plot_surface(w0, w1, hinge_z, rstride=5, cstride=5, cmap=cm.coolwarm,
            linewidth=0, antialiased=False)
    ax.set_xlabel('w0')
    ax.set_ylabel('w1')
    ax.set_zlim(0, 1200)
    ax.set_title('Linear SVM Learning Surface')
    
    ax = fig.add_subplot(1, 2, 2, projection='3d')
    #ax = fig.gca(projection='3d')
    surf = ax.plot_surface(w0, w1, log_z, rstride=5, cstride=5, cmap=cm.coolwarm,
            linewidth=0, antialiased=False)
    ax.set_xlabel('w0')
    ax.set_ylabel('w1')
    ax.set_zlim(0, 1200)
    ax.set_title('Logistic Regression Learning Surface')
    
sep_surface()
````

"In mathematical optimization, statistics, decision theory and machine learning, a loss function or cost function is a function that maps an event or values of one or more variables onto a real number intuitively representing some "cost" associated with the event. An optimization problem seeks to minimize a loss function. An objective function is either a loss function or its negative (sometimes called a reward function, a profit function, a utility function, a fitness function, etc.), in which case it is to be maximized." https://en.wikipedia.org/wiki/Loss_function   

The Loss function consists of:
    – loss term ($L(m_i(w)$, expressed in terms of the margin of each training example) and
    – regularization term ($R(w)$ expressed as a function of the model complexity)

\begin{equation}
J(W)  =  \underbrace{\sum_i L(m_i(w))}_\text{Loss} + \underbrace{\lambda R(w)}_\text{Regularization}
\end{equation}


The linear SVM learning algorithm and an L2 penalized logistic regesssion may provide similar results, but the underlying functions are quite different. The SVM uses a hinge loss function, where as LR uses logloss - the differences can be seen in the graph above. While SVMs and perceprtons are deterministic in nature, logistic regression aims to maximize the posterior class probability.

In the case of binary classification, both the SVM and the perceptron will try to find the decision boundry between positive and negative exmaples. The perceptron has a tendency to be less generalizable than the SVM. It is a simpler model and while it finds the separating hyperplane, it doesn't necessarily find the "best" hyperplane. The SVM on the other hand, tries to maximize the margin between the closest support vectors.

The linear SVM algorithm finds an optimally separating hyperplane to maximize the margin. The perceptron takes repeated passes through the data and self-corrects each time it makes a mistake, eventually converging on a hyperplane that separates the data. Perceptron training stops once a weight vector is found that correctly classifies each example.  So perceptron will not update coeffcients when data has already been seperated, which means that it will in general not find the optimally sperating hyperplane to maximize the margin.

In addition, a soft SVM would be better equipped to handle linearly inseparable data because slack variables allow for imperfect partitioning of data. The perceptron does not have this flexibility and would have difficulty converging or would not separate the data very well.

SVMs have a different objective function to the perceptron  and gradient descent only stops updating when the change to the weight vector is small.  These are approximate algorithms so convergence is not guaranteed but you should never the less get a good separating hyperplane with largest margin (if learning a Hard SVM).

## Logistic Regressiion Penalized Loss Functions

### Ridge regularization - L2
If there exists more than one solution to the minimization problem, then this problem is referred to as ill-posed and can lead to over-fitting. L2 (Tikhonov regularization, named for Andrey Tikhonov), is the most commonly used method of regularization of ill-posed problems. Regularization has the effect of "shrinking" the coefficients when the predictors are highly correlated.

#### Objective Function with ridge regularization term

\begin{equation}
\begin{split}
arg min_w~L(w) = \underbrace{\frac{1}{n} \sum_{i = 0}^{n-1} log(1 + e^{-y.w^T.x_i})}_\text{Loss} + \underbrace{\lambda ||w||^2}_\text{Penalty} \\
y \in \{-1, 1\}
\end{split}
\end{equation}

#### Gradient Descent

\begin{equation}
\frac{\partial L}{\partial w} = \frac{1}{n} \sum_{i = 0}^{n-1} (\frac{1}{1+e^{-y.w^T.x_i}} -1 ).y_i.x_i + \lambda w
\end{equation}

### Lasso regularization - L1
Like Ridge, Lasso shrinks the coeeficients of highly correlated predictros. Unlike Ridge however, Lasso can set coefficients to zero, effectively reducing the number of features. It can be seen in the figure above, that for some features, the closest point to the unconstrained minimum which is within our <a href="#budget">"budget"</a>, equals 0.

### ElasticNet - L1 + L2
In statistics and, in particular, in the fitting of linear or logistic regression models, the elastic net is a regularized regression method that linearly combines the L1 and L2 penalties of the lasso and ridge methods. "The elastic net often outperforms the lasso, while enjoying a similar sparsity of representation. ### Bayesian interpretation - L1 and L2 priors

<!-- <img src="https://upload.wikimedia.org/wikipedia/en/f/fd/L1_and_L2_balls.jpg" style="width:90%;"/> -->
Just as ridge regression can be interpreted as linear regression for which the coefficients have been assigned normal prior distributions, lasso can be interpreted as linear regression for which the coefficients have Laplace prior distributions. The Laplace distribution is sharply peaked at zero (its first derivative is discontinuous) and it concentrates its probability mass closer to zero than does the normal distribution. This provides an alternative explanation of why lasso tends to set some coefficients to zero, while ridge regression does not.
Ref: https://en.wikipedia.org/wiki/Lasso_(statistics)#Bayesian_interpretation

Without regularization, the prior belief is that all regressors are equally important. In reality this is rarely the case. Some variables will be correlated and thus have little to no bearing on the outcome. "A reasonable approach to quantifying this belief is to place zero-mean normal priors on the regression coefficients to indicate our preference for smaller values. This choice of prior distribution gives rise to the technique of ridge regression."[1] Giving preference to smaller values necessarily penalizes outliers.

In cases where there are many more features than examples, L1 regularization comes in particulary handy in setting many of these features to zero. For supervised learning problems where there is a very large number n of input features, but where there may be a small subset—say r << n of them that is sufficient to learn the target concept well, Andrew Y. Ng shows that L1 normalization works well.[2]
***
We can view the L1 and L2 as priors because their values are based on the weights from the previous iteration of logistic regression. Both regularizers function as constraints for the values that the weights can take, and they bias the model toward the prior weights depending on the size of the regularization parameter (larger parameter = more bias). The L1 regularizer can be thought of as the log of a Laplace prior, and the L2 can be thought of as log of a Gaussian prior. Because L1 is not differentiable, it can be used for feature selection.

````python
#######################################################################
# Home Grown implementation
#######################################################################
from __future__ import division
from collections import namedtuple
import numpy as np
import copy


#######################################################################
# @input:   
# -1,-0.284348239821,0.812776581897
# @output normalized points:  
# [Point(x=[-0.33022258,  0.94390309,  1.], y=-1.0)]
#######################################################################
Point = namedtuple('Point', 'x y')
def readPoint(line):
    d = line.split(',')
    x = [float(i) for i in d[1:]]
    x.append(1.0)  # bias term
    x = normalize(x)
    return Point(x, float(d[0]))
#######################################################################



#######################################################################
# Weight the above training dataset as follows: 
# Weight each example using the inverse vector length (Euclidean norm):
#######################################################################
def normalize(V):
    # only normalize the coefficients, not intercept
    intercept = V[-1]
    coefficients = V[:-1]
    d = np.sqrt(sum(np.array(coefficients)**2))
    coefficients = [c/d for c in coefficients]
    return np.hstack([np.array(coefficients), np.array(intercept)])
#######################################################################
````

````python
#######################################################################
# gradient descent (with NO stochasticity!)
# Objective Function:
# minw   λ/2   w’w   +   1/m Σi log(1+exp(yi(w’xi – b)))   
# gradient function A = -y*(1-1/(1+exp(yi(w’x – b))))*x
# gradient function B = (1/(1+exp(-y*(w’x))))*y*x
#######################################################################

def logisticRegressionGD_homeGrown(data, \
                                   wInitial=None, \
                                   learningRate=0.05, \
                                   iterations=50, \
                                   regParam=0.01, \
                                   regType="Lasso"):
    featureLen = len(data.take(1)[0].x)
    n = data.count()
    
    if wInitial is None:
        w = np.random.normal(size=featureLen)
    else:
        w = wInitial
    for i in range(iterations):
        wBroadcast = sc.broadcast(w)
        
        gradient = data.map(lambda p: -(1 - 1 / (1 + np.exp(-p.y*np.dot(wBroadcast.value, p.x)))) \
                                    * p.y \
                                    * np.array(p.x)) \
                                    .reduce(lambda a, b: a + b)/n

        if regType == "Ridge":
            wReg = w * 1 ## the derivative of w^2 is 2*w, but the constant doesn't matter. 
            wReg[-1] = 0 #last value of weight vector is bias term (intercept), ignored in regularization
        elif regType == "Lasso":
            wReg = w
            wReg[-1] = 0 #last value of weight vector is bias term (intercept), ignored in regularization
            wReg = (wReg>0).astype(int) * 2-1 ## the derivative of |w| is {-1,1} 
        else:
            wReg = np.zeros(n)
        
        gradient = gradient + regParam * wReg  #gradient:  GD of Sqaured Error+ GD of regularized term 
        
        w = w - learningRate * gradient
        
    return w
````

## SVM Functions
### Hinge loss with ridge regularization 
Hinge loss is used for "maximum-margin" classification, most notably for support vector machines (SVMs). For an intended output t = ±1 and a classifier score y, the hinge loss of the prediction y is defined as
\begin{equation}
l_{\text{hinge}} = \max(0,1-y\ \boldsymbol{x}\cdot\boldsymbol{w})
\end{equation}

#### Regularized Hinge Losss
\begin{equation}
argmin_w  \frac{1}{n}\sum_i(1   –    y_i(w’x_i – b))_+ + \lambda ||\vec w||^2
\end{equation}


#### Gradient Descent
Because the hinge loss is not differentiable, a sub-gradient is computed

$$λw             \  if \      y_i(w’x_i – b))  > 1    \ - correctly \ classified$$
$$λw  + y_i x_i \    Otherwise                     \ - incorrectly \ classified$$

***
For **HARD SVM** we are concerned only with the regularization component $\frac{2}{||\vec w||}$ and not with the loss, where the regularization component is the length of the weight vector. In the case of **HARD SVM**, a penalty is given if margin < 1, and there is zero tolerance for mistakes. This means that the data must be linearly separable for the Hard SVM algorithm to converge. 

This restriction is relaxed in the **SOFT SVM** by introducing a "slack" component:

$$C\sum_{i=1}^L\xi_i$$

The soft margin loss function 

\begin{equation}
argmin_w = \frac{1}{2}||\vec w||^2 + C\sum_{i=1}^L\xi_i
\end{equation}

subject to

\begin{equation}
y_i(\left\langle w,x_i\right\rangle + b) \geq 1 - \xi_i  \\
\end{equation}

allows for some mistakes. In the MMlib implementation the SOFT SVM can be tuned using the regParam parameter.

````python
#gradient descent (and with NO stochasticity!)
# Objective Function
# minw   λ/2   w’w   +   1/m Σi(1   –    yi(w’xi – b))+
# gradient 
#   λw               if      yi(w’xi – b))  > 1    # correctly classified
#   λw  + yi xi      Otherwise                    # incorrectly classified
#--------------------------------
#Wt+1  = wt + average(gradient)
#Wt+1  = wt + average(regularization + hinge loss)

def SVM_GD_SPARK(data,w=None,eta=0.01,iter_num=10,regPara=0.01,stopCriteria=0.0001):
    n = data.count()
    
    if w==None:
        w = np.random.normal(size=3)
           
    for i in range(iter_num):       #label * margin
        wBroadcast = sc.broadcast(w)
        sv = data.filter(lambda p: p.y * np.dot(wBroadcast.value,np.array(p.x)) < 1 )  # Support vector? with label*margin<1
#         if i % iter_num == 0: # for debugging purposes
#             print "values, weights", sv.take(1), wBroadcast.value
#             print "Number of support vectors",sv.count()

        if sv.isEmpty(): # converged as no more updates possible
            print "converged as no more updates possible"
            break       # hinge loss compoent of gradient y*x and sum up
            
        g = -sv.map(lambda p: p.y * np.array(p.x)).reduce(lambda x,y:x+y)/n #gradient: avg hinge loss
        
        wreg = w*1   # temp copy of weight vector
        wreg[-1] = 0 #last value of weight vector is bias term; ignore in regularization
        
        wdelta = eta*(g+regPara*wreg)  #gradient: hinge loss + regularized term 
        w = w - wdelta
    return w
````
