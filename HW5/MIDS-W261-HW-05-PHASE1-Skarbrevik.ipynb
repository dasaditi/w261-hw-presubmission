## HW5.0  data warehouse; star schema

- What is a data warehouse? What is a Star schema? When is it used?

1) A data warehouse is a repository of data. Traditionally in the form of relational databases but are now storing more semi-structured and unstructured forms of data. Traditionally two forms of databases: OLTP and OLAP. OLAP systems are meant for relatively low volume transactions, where latency isn't much of an issue. Queries may often be very complex. OLTP systems on the other hand focus on high volumes of transactions with relatively simple operations.

2) A Star schema is a type of schema used to develop data warehouses . In this schema, a central fact table references other branched tables that contain different dimensions.

- In the database world What is 3NF? Does machine learning use data in 3NF? If so why? 
- In what form does ML consume data?
- Why would one use log files that are denormalized?

3NF refers to third normalized form of a database. This means that related data is stored in physically separate logical databases. The advantage of this is that writing data in this schema can be much easier. Machine Learning does not use 3NF because it creates too much computational strain to be putting normalized data together to perform complex calculations.

Machine learning needs denormalized forms of data. This means that all relevant data for an algorithm will be conveniently located.

````python
%%writefile MapperSideInnerJoin.py
#!/usr/bin/env python

import mrjob
from mrjob.job import MRJob
from mrjob.step import MRStep
#from mrjob.compat import get_jobconf_value
 
class innerjoin(MRJob):
    
    def mapper_init(self):
        self.countries = {}
        self.countries_onRight = {}
        self.left = 0
        self.right = 0
        self.inner = 0
        
        with open("Countries.dat","r") as file:
            for line in file:
                line = line.strip()
                data = line.split("|")
                self.countries[data[1]] = data[0]
            
            
    def mapper(self, _, line):
        line = line.strip()
        data = line.split("|")
        key = data[2]
        
        if key in self.countries.keys():
            self.inner += 1
            yield None, "inner,"+key+","+self.countries[key]+","+data[0]+","+data[1]
            self.left += 1
            yield None, "left,"+key+","+self.countries[key]+","+data[0]+","+data[1]
            self.right += 1
            yield None, "right,"+key+","+self.countries[key]+","+data[0]+","+data[1]
            self.countries_onRight[key] = 1
        else:
            self.right += 1
            yield None, "right,"+key+","+data[0]+","+data[1]
        
    
    def mapper_final(self):
        for key in self.countries.keys():
            if key not in self.countries_onRight.keys():
                self.left += 1
                yield None, "left,"+key+","+self.countries[key]+",NA,NA"
        yield None, ""
        yield None, "left join produces a table with "+str(self.left)+" rows."
        yield None, "inner join produces a table with "+str(self.inner)+" rows."
        yield None, "right join produces a table with "+str(self.right)+" rows."

            

if __name__ == '__main__':
    innerjoin.run()
````

## HW5.2.1 (OPTIONAL) Almost stateless reducer-side join

The following MRJob code, implements a reduce-side join for an inner join. The reducer is almost stateless, i.e., uses as little memory as possible. Use the tables from HW5.2 for this HW and join based on the country code (third column of the transactions table and the second column of the Countries table perform. Perform  an left, right, inner joins using the code provided below and report the number of rows resulting from:

- (2) Right joining Table Left with Table Right
- (3) Inner joining Table Left with Table Right

__Some notes on the code__ 
Here, the mapper receives its set of input splits either from the transaction table or from the countries table and makes the appropriate transformations: splitting the line into fields, and emitting a key/value. The key is the join key - in this case, the country code field of both sets of records. The mapper knows which file and type of record it is receiving based on the length of the fields. The records it emits contain the join field as the key, which acts as the partitioning key; We use the SORT_VALUES option, which ensures the values are sorted as well. Then, we employ a trick to ensure that for each join key, country records are seen always before transaction records. We achieve this by adding an arbitrary key to the front of the value: 'A' for countries, 'B' for customers. This makes countries sort before customers for each and every join/partition key. After that trick, the join is simply a matter of storing countries ('A' records) and crossing this array with each customer record.

````python
import sys, os, re
from mrjob.job import MRJob

class MRJoin(MRJob):

  # Performs secondary sort
  SORT_VALUES = True

  def mapper(self, _, line):
    splits = line.rstrip("\n").split("|")

    if len(splits) == 2: # country data
      symbol = 'A' # make country sort before transaction data
      country2digit = splits[1]
      yield country2digit, [symbol, splits]
    else: # person data
      symbol = 'B'
      country2digit = splits[2]
      yield country2digit, [symbol, splits]

  def reducer(self, key, values):
    countries = [] # should come first, as they are sorted on artificia key 'A'
    for value in values:
      if value[0] == 'A':
        countries.append(value)
      if value[0] == 'B':
        for country in countries:
          yield key, country[1:] + value[1:]

if __name__ == '__main__':
  MRJoin.run()
````

````python
%%writefile buildStripes.py
#! /opt/anaconda/bin/python 
# -*- coding: utf-8 -*-

from __future__ import division
import re
import mrjob
import json
from mrjob.protocol import RawProtocol
from mrjob.job import MRJob
from mrjob.step import MRStep

class MRbuildStripes(MRJob):
    
    def mapper_init(self):
        self.WORD_RE = re.compile(r"[\w']+")
    
    def mapper(self, _, line):
        self.increment_counter('group', 'Num_mapper_calls', 1)
        line = line.strip()
        data = line.split("\t")
        ngram = self.WORD_RE.findall(data[0])
        for i in range(len(ngram)):
            ngram[i] = ngram[i].lower()

        count = data[1]
        H = {}
        
        for word in ngram:
            # enter new stripe if necessary
            if word not in H.keys():
                H[word] = {}
                for coWord in ngram:
                    if coWord == word:
                        continue
                    elif coWord not in H[word]:
                        H[word][coWord] = int(count)
                    else:
                        H[word][coWord] += int(count)
                continue
            # update stripes already in existance
            for coWord in ngram:
                if coWord == word:
                    continue
                elif coWord not in H[word]:
                    H[word][coWord] = int(count)
                else:
                    H[word][coWord] += int(count)
        # emit stripes
        for keys in H.keys():
            yield keys, H[keys]
        
        
    def reducer(self, key, value):
        self.increment_counter('group', 'Num_reducer_calls', 1)
        H = list(value)
        tmp = {}
        # aggregate stripe data
        for i in range(len(H)):
            for words in H[i].keys():
                if words in tmp:
                    tmp[words] += int(H[i][words])
                else:
                    tmp[words] = int(H[i][words])
        yield key, tmp
                
        
    def steps(self):
        return [
                self.mr(mapper_init = self.mapper_init,
                      mapper = self.mapper,
                      reducer = self.reducer)
        ]

if __name__ == '__main__':
    MRbuildStripes.run()
````

````python
%%writefile invertedIndex.py
#! /opt/anaconda/bin/python 
# -*- coding: utf-8 -*-


from __future__ import division
import collections
import re
import json
import math
import numpy as np
import itertools
import mrjob
from mrjob.protocol import RawProtocol
from mrjob.job import MRJob
from mrjob.step import MRStep

class MRinvertedIndex(MRJob):
    
    MRJob.SORT_VALUES = True 
     
    def mapper(self,_,line):

        line = line.strip()
        doc_id, words = line.split("\t")
        stripe = json.loads(words)

        _len = len(stripe)
        for words in stripe.keys():
            # Store the length of the document to use with JACCARD (|A| + |B|)
            yield words, (doc_id, _len)
        
        
        
    def reducer(self,key,value):

        d = collections.defaultdict(list)
        for v in value:
            d[key].append(v)
        yield key,d[key]
        
    def steps(self):

        return [MRStep(
                    mapper=self.mapper,
                    reducer=self.reducer)
                ]
        
if __name__ == '__main__':
    MRinvertedIndex.run() 
````

````python
%%writefile similarity.py
#! /opt/anaconda/bin/python 
# -*- coding: utf-8 -*-

from __future__ import division
import collections
import re
import json
import math
import numpy as np
import itertools
import mrjob
from mrjob.protocol import RawProtocol
from mrjob.job import MRJob
from mrjob.step import MRStep

class MRsimilarity(MRJob):
  
    MRJob.SORT_VALUES = True 
    def steps(self):

        JOBCONF_STEP1 = {}
        JOBCONF_STEP2 = { 
          ######### IMPORTANT: THIS WILL HAVE NO EFFECT IN -r local MODE. MUST USE -r hadoop FOR SORTING #############
            'mapreduce.job.output.key.comparator.class': 'org.apache.hadoop.mapred.lib.KeyFieldBasedComparator',
            'mapreduce.partition.keycomparator.options':'-k1,1nr',
        }
        return [MRStep(jobconf=JOBCONF_STEP1,
                    mapper=self.mapper_pair_sim,
                    reducer=self.reducer_pair_sim)
                ,
                MRStep(jobconf=JOBCONF_STEP2,
                    mapper=None,   
                    reducer=self.reducer_sort)
                ]
            
    def mapper_pair_sim(self,_,line):
        line = line.strip()
        index,posting = line.split("\t")
        posting = json.loads(posting)
        
        '''
        @input: lines (postings) from inverted index
         "blue" [["DocA", 4], ["DocC", 4], ["DocE", 3]]
        
        @output: pairs of doc and doc_length, count the number of pairs
         make complex key and count of 1 as value:
         DocA.4.DocB.4, 1
         DocA.4.DocC.4, 1
         DocB.4.DocC.4, 1
        '''
        
        X = map(lambda x: x[0]+"."+str(x[1]) , posting)
        
        # taking advantage of symetry, output only (a,b), but not (b,a)
        for subset in itertools.combinations(sorted(set(X)), 2):
            yield subset[0]+"."+subset[1], 1



    def reducer_pair_sim(self,key,value):
        Doc1, Doc1_len, Doc2, Doc2_len = key.split(".")
        t = sum(value)
        
        jaccard = t / ( int(Doc1_len) + int(Doc2_len) - t )
        
       
        yield jaccard, (Doc1+" - "+Doc2)
    
    
    def reducer_sort(self,key,value):
        for v in value:
            yield key,v
  
if __name__ == '__main__':
    MRsimilarity.run()
````
