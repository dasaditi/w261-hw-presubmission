<h2>Class to initialize the graph<a name="GraphInit"></a><a style="float:right;" href="#top">^ top</a></h2>

````python
%%writefile initGraph.py
#!~/anaconda2/bin/python
# -*- coding: utf-8 -*-
from __future__ import division
import mrjob
from mrjob.job import MRJob
from mrjob.step import MRStep
import json
import ast
import re
import numpy as np
class initGraph(MRJob):
    
    def configure_options(self):
        super(initGraph, self).configure_options()
        self.add_passthrough_option(
            '--startNode', type='str', default='1', help='Select the start node')
        self.add_passthrough_option(
            '--weighted', type='str', default='F', help='Is this a weighted graph?')
        self.add_passthrough_option(
            '--distance', type='str', default='shortest', help='Longest or shortest distance?')


        
    def mapper1(self, _, line):
        line = line.strip()
        key,value = line.split("\t")
        adjList = ast.literal_eval(value)
        
        # key -> nodeId
        # value -> [state<Q|U|V>, neighbors, distance, path]
        # by placing the state first (Q comes before U and V), 
        # we take advantage of secondary sorting in the SSSP job
        # and avoid crazy if/else logic.
        
        #  "for the sake of this assignment, treat this an unweighted network,
        #  and set all weights to 1 upon data input!!"
        if self.options.weighted == "F":
            for k in adjList:
                adjList[k] = 1
        
        
        if key == self.options.startNode:
            yield key, ("Q",adjList,0,key)
        else:    
            if self.options.distance == "longest":
                yield key, ("U",adjList,0,key)
            else:    
                yield key, ("U",adjList,float("inf"),key)
    
    
            
    def steps(self):
        return [MRStep(
                mapper=self.mapper1,
                reducer=None)
               ]

if __name__ == '__main__':
    initGraph.run()
````

<h2>Class to calculate Single Source Shortest Path<a name="sssp"></a><a style="float:right;" href="#top">^ top</a></h2>

````python
%%writefile SingleSourceShortestPath.py
#!~/anaconda2/bin/python
# -*- coding: utf-8 -*-
from __future__ import division
import mrjob
from mrjob.job import MRJob
from mrjob.step import MRStep
import json
import ast
import re
import sys
import boto3
import numpy as np
class SingleSourceShortestPath(MRJob):
    
    MRJob.SORT_VALUES = True
    def configure_options(self):
        super(SingleSourceShortestPath, self).configure_options()
        self.add_passthrough_option(
            '--targetNode', type='str', default=536, help='Select the taget node')

    def mapper(self, _, line):
      
        line = line.strip()
        key, value = line.split("\t")
        key = key.replace('"','') # remove extra quotation marks
        value = json.loads(value) 
        
        status = value[0]
        neighbors = value[1]
        distance = value[2]
        path = value[3]
       
        if status == "Q":
            # put neighbors in Q mode and update path length by incrementing path length of N
            for neighbor in neighbors:
                yield neighbor, ("Q", {}, distance + int(neighbors[neighbor]), str(path)+" -> "+str(neighbor))

            # Update status of node N to Visited
            status = "V"

        yield key, (status, neighbors, distance, path)
        
    
    def reducer(self, key, values):
        _neighbors = {}
        _distance = float("inf")
        _status = ""
        _path = ""
        
        # assume no outbound links if all keys are in Q status
        # in other words, there are no keys in U or V status for this node
        no_outbound_links = 0
        
        for v in values:
            neighbors = v[1]
            distance = v[2]
            status = v[0]
            path = v[3]
            
            # !! Because of secondary sorting, Q comes before U and V
            if status == "Q":
                # only for unweighted graphs. For weighted graphs, can't stop just 
                # because a node has been visited, since there may be other shorter paths
                if key == self.options.targetNode:
                    self.increment_counter('Q', 'stop', 1)
                    s3 = boto3.resource('s3')
                    s3.Bucket('ucb261-hw7').put_object(Key='tmp/path_to_'+str(key)+'.txt', Body=json.dumps([path,distance]))
                    
                self.increment_counter('Q', 'Q_count', 1)
                
                if distance < _distance:
                    _distance = distance
                    _path = path                    
            
            else:
                no_outbound_links += 1
                
                if _distance < distance:
                    status = "Q"
                    distance = _distance
                    path = _path            
                
                yield key, (status, neighbors, distance, path)
                
                
                
        if no_outbound_links == 0:
            yield key, ("V", neighbors, distance, path)

     
    def steps(self):
             
        JOBCONF = {
            "mapreduce.map.memory.mb": "8192",                                    
            "mapreduce.map.java.opts": "-Xmx6144m",                                 
            "mapreduce.reduce.memory.mb": "8192",                                         
            "mapreduce.reduce.java.opts": "-Xmx6144m"                                           
        }
        return [MRStep(jobconf=JOBCONF,
                mapper=self.mapper,
                reducer=self.reducer
                )
               ]

if __name__ == '__main__':
    SingleSourceShortestPath.run()
````

<h2>EDA Class to calculate stats<a name="eda"></a><a style="float:right;" href="#top">^ top</a></h2>

````python
%%writefile wiki_eda.py
#!~/anaconda2/bin/python
# -*- coding: utf-8 -*-
from __future__ import division
import mrjob
from mrjob.job import MRJob
from mrjob.step import MRStep
import json
import ast
import re
import sys
import numpy as np
class wiki_eda(MRJob):
    
    MRJob.SORT_VALUES = True
    '''
    Sample input:
    1	{'2': 1, '6': 1}
    2	{'1': 1, '3': 1, '4': 1}
    3	{'2': 1, '4': 1}
    4	{'2': 1, '5': 1}
    5	{'1': 1, '2': 1, '4': 1}
    '''
    def __init__(self, *args, **kwargs):
        super(wiki_eda, self).__init__(*args, **kwargs)
        self.vocab = 0
        
    def mapper(self, _, line):
      
        line = line.strip()
        node, edges = line.split("\t")
        edges = ast.literal_eval(edges)
        
        # Account for all nodes including nodes with no out-links
        yield "*!"+str(node), 1
        for e in edges:
            yield "*!"+str(e), 1
        
        # emit length of edges as key to generate a distribution graph
        # kinda like a group by edge count
        yield len(edges), 1
        
        # emit counts: (length of adjacency list) and (1) to count number of nodes
        yield "*counts", [len(edges), 1]
       
    def reducer(self, key, value):
        total_edges = 0
        total_nodes = 0
        if str(key).startswith("*!"):
            self.vocab += 1 
        elif key == "*counts":
            for v in value: 
                total_edges += v[0]
                total_nodes += v[1]
            
            # I want these stats to sort to the bottom of the file, hence the maxint    
            yield "Average degree: "+str(total_edges/self.vocab), sys.maxint
            yield "Number of nodes: "+str(self.vocab), sys.maxint
            yield "Number of out-links: "+str(total_edges), sys.maxint
            
        else:
            # edgecount, number of nodes with this edgecount
            yield int(key), sum(value)
    
    def identity_reducer(self,key,value):
        for v in value:
            yield key, v
    
    def steps(self): 
        JOBCONF1 = {
              'mapred.reduce.tasks': '1'
        }
        JOBCONF2 = {
              'mapred.output.key.comparator.class':'org.apache.hadoop.mapred.lib.KeyFieldBasedComparator',
              'mapred.text.key.comparator.options': '-k1n -k2nr',
              'mapred.reduce.tasks': '1'
              # this needs to have one reducer, since we are not doing any custom partitioning,
              # plus the number of records is only as many as there are unique edge counts 
              # so no 'need' for more reducers
        }
        
        return [MRStep(jobconf=JOBCONF1,
                mapper=self.mapper,
                reducer=self.reducer
                ),
                MRStep(jobconf=JOBCONF2,
                mapper=None,
                reducer=self.identity_reducer       
                
                )
               ]

if __name__ == '__main__':
    wiki_eda.run()
````

<h2>Join NLTK indices (in memory) with output from previous step streamed in<a name="joinNLTK"></a><a style="float:right;" href="#top">^ top</a></h2>

````python
%%writefile performJoins.py
#!~/anaconda2/bin/python
# -*- coding: utf-8 -*-

import mrjob
from mrjob.protocol import RawValueProtocol
from mrjob.job import MRJob
from mrjob.step import MRStep
import csv
import json
import re

import sys

targetNode = sys.argv[1]

class performJoins(MRJob):
    OUTPUT_PROTOCOL = RawValueProtocol
    
    def configure_options(self):
        super(performJoins, self).configure_options()
        self.add_passthrough_option(
            '--targetNode', type='str', default=536, help='Select the taget node')

    
    def steps(self):
        JOBCONF_STEP1 = {    
            "mapred.reduce.tasks":1
            # ^^ Since we will end up with a small file
        }
        return [MRStep(jobconf=JOBCONF_STEP1,
            mapper_init = self.mapper_init,
            mapper = self.mapper,
            reducer = self.reducer    
        )]

    def mapper_init(self):
        self.IDX = {}
        f = open("indices.txt","r")
        for line in f:
            line = line.strip()
            word, idx = line.split("\t")
            self.IDX[idx] = word

    def mapper(self,_,line):
        line = line.strip()
        key,value = line.split("\t")
        key = key.replace('"','')
        value = json.loads(value)
        
        if key == self.options.targetNode:
            
            distance = value[2]
            path = value[3]

            path = path.strip().split(" -> ")
            words = []

            for p in path:
                words.append(self.IDX[p])

            yield None, "The shortest distance from '%s' to '%s' is: %s"%(words[0], words[-1], str(distance))
            yield None, "The path from '%s' to '%s' is: %s: "%(words[0], words[-1], ' -> '.join(words))
            
    def reducer(self, key, value):
        for v in value:
            yield key, v

if __name__ == '__main__':
    performJoins.run()
````

````python
################################################################
# Run on EMR
###############################################################
from __future__ import division
import time
import json
from SingleSourceShortestPath import SingleSourceShortestPath
from initGraph import initGraph


cluster_id = 'j-2KLOG1841BVQU'

startNode = "4792405"
targetNode = "9779957"

# 4792405  "Friedlander-Iwaniec theorem"
# "9779957" "Ordinary Love (U2 song)"
# "6176135" Ireland 
# "13466359" Berkeley 

# !aws s3 rm --recursive s3://ucb261-hw7/alt_wiki__0
# !aws s3 rm --recursive s3://ucb261-hw7/wiki__0      

    
input_dir = 's3://ucb261-hw7/alt_wiki__'
output_dir = input_dir


iteration = 0 

###############################################################
# JOB 1  -  make the graph structure and mark start node   
###############################################################

mr_job_init = initGraph(args=['-r','emr',
                              's3://ucb-mids-mls-networks/wikipedia/all-pages-indexed-out.txt',
                              '--cluster-id='+cluster_id,
                              '--output-dir='+input_dir+str(iteration),
                              '--startNode='+startNode,
                              '--weighted=F'])

with mr_job_init.make_runner() as runner: 
        runner.run()
        print "written graph" 

###############################################################               
# JOB 2  -  iterate until convergence
###############################################################

start = time.time()


while(1):
    end = time.time()
    print "iteration: ",iteration, ", time: ", (end-start)/60, " minutes"
    
    mr_job = SingleSourceShortestPath(args=[
                             '-r','emr',
                             input_dir+str(iteration),
                             '--cluster-id='+cluster_id,
                             '--output-dir='+output_dir+str(iteration+1),
                             '--targetNode='+targetNode
                             ])
    
    !echo "Wait 5.0 sec"
    !sleep 5
    
    iteration += 1
    print "iteration: ",iteration
    
    with mr_job.make_runner() as runner: 
        runner.run()
        
        if "Q" in runner.counters()[0]:
            print runner.counters()[0]['Q']
        
        # terminate when there are no output nodes that are frontier (i.e., in Q state) or we've reached the target
        if "Q" not in runner.counters()[0] or ('stop' in runner.counters()[0]['Q'] \
                                               and runner.counters()[0]['Q']['stop'] > 0):   
            print "\n--- finished %s ---\n" % iteration    
            break
    
end = time.time()
print "Total time: ", (end-start)/60, " minutes"
````

<h2>Join Wikipedia indices with path<a name="joinWIKI"></a><a style="float:right;" href="#top">^ top</a></h2>
<p>Get  page titles
<br>Stream in the wikipedia index, keep path in memory</p>

````python
%%writefile joinWithIndex.py
#!~/anaconda2/bin/python
# -*- coding: utf-8 -*-

import mrjob
from mrjob.protocol import RawValueProtocol
from mrjob.job import MRJob
from mrjob.step import MRStep
import csv
import json
import re
import sys

targetNode = sys.argv[1]

class joinWithIndex(MRJob):
#     OUTPUT_PROTOCOL = RawValueProtocol
    
    def configure_options(self):
        super(joinWithIndex, self).configure_options()
        self.add_passthrough_option(
            '--targetNode', type='str', default=536, help='Select the target node')

    
    def steps(self):
        JOBCONF_STEP1 = {    
            "mapred.reduce.tasks":1
            # ^^ Since we will end up with a small file
        }
        return [MRStep(jobconf=JOBCONF_STEP1,
            mapper_init = self.mapper_init,
            mapper = self.mapper,
            reducer = self.reducer    
        )]

    def mapper_init(self):
        self.IDX = {}
        f = open("path_to_"+self.options.targetNode+".txt","r")
        # input: ["6176135 -> 11607791 -> 13466359", 2]
        for line in f:
            line = line.strip()
            line = json.loads(line)
            path = line[0].split(" -> ")
            
            for i,idx in enumerate(path):
                self.IDX[idx] = i

    def mapper(self,_,line):
        line = line.strip()
        title, indx, inDeg, outDeg = line.split("\t")
        
        if indx in self.IDX:
            yield self.IDX[indx], title
            
    def reducer(self, key, value):
        for v in value:
            yield key, v

if __name__ == '__main__':
    joinWithIndex.run()
````

````python
%%writefile findIndex.py
#!~/anaconda2/bin/python
# -*- coding: utf-8 -*-

import mrjob
from mrjob.protocol import RawValueProtocol
from mrjob.job import MRJob
from mrjob.step import MRStep

class findIndex(MRJob): 
        
    def configure_options(self):
        super(findIndex, self).configure_options()
        self.add_passthrough_option(
            '--sourceNode', type='str', default="Ireland", help='Select the source node')
        self.add_passthrough_option(
            '--targetNode', type='str', default="Berkeley", help='Select the taget node')

    def steps(self):
        JOBCONF_STEP1 = {    
            "mapred.reduce.tasks":1
            # ^^ Since we will end up with a small file
        }
        return [MRStep(jobconf=JOBCONF_STEP1,
            mapper = self.mapper,
            reducer = self.reducer    
        )]

    def mapper(self,_,line):
        line = line.strip()
        title, indx, inDeg, outDeg = line.split("\t")
        
        if title.lower().find(self.options.sourceNode.lower()) > -1:
            yield indx, title

        if title.lower().find(self.options.targetNode.lower()) > -1:
            yield indx, title
            
    def reducer(self, key, value):
        for v in value:
            try:
                yield key, unicode(v, errors='ignore')
            except:
                yield None,"bad enconding"

if __name__ == '__main__':
    findIndex.run()
````

https://en.wikipedia.org/wiki/Longest_path_problem   
<p>In graph theory and theoretical computer science, the longest path problem is the problem of finding a <strong>simple</strong> path of maximum length in a given graph. A path is called simple if it does not have any repeated vertices; the length of a path may either be measured by its number of edges, or (in weighted graphs) by the sum of the weights of its edges. In contrast to the shortest path problem, which can be solved in polynomial time in graphs without negative-weight cycles, the longest path problem is NP-hard, meaning that it cannot be solved in polynomial time for arbitrary graphs unless P = NP. Stronger hardness results are also known showing that it is difficult to approximate.</p>
<p>The Longest Path Problem (LPP) for a cyclic graph is a NP-complete whereas finding the Shortest Path Problem (SPP) is a problem known to be in P. The proof for the NP-completeness of LPP consists of a reduction from the Hamiltonian Circuit (HC) problem which is already known to be NP-complete.</p>

A combiner can help with reducing the network traffic and lessening the computation load on the reducers.

The order inversion pattern exploits the sorting phase of MapReduce to push data needed for calculations to the reducer ahead of the data that will be manipulated.

Using the order inversion pattern for values can simplify the reducer code. By using secondary sort, values with 'Q' status will arrive before V or U, thus making it easier to extract the minimum distance in the case of a weighted graph, and make a decision about resetting a visited node to Q status. We want to pick the path with the minimum distance from among the nodes in the Q state before comparing this result to a possible node in a V state. Thus having all Q nodes arrive first simplifies this process.
