## MRJob Classes for Breadth-First Search / Shortest Path

````python
%%writefile ShortestPathExpand.py
# ! /opt/anaconda/bin/python 
# -*- coding: utf-8 -*-

from __future__ import division
import re
import mrjob
import json
from mrjob.protocol import RawProtocol
from mrjob.job import MRJob
from mrjob.step import MRStep


# This is step 2 / 2 in finding the path distances from one node in a graph to another node
# In the previous step we assigned initial priority queue values to every node
# In this step we will iterate over the priority queue until all nodes have been processed
# and all distances have been calculated.

# Explanation of priority queue formatting seen in code below:
# Q means these nodes are ready for expansion.
# V means node has been visited and doesn't need to be processed.
# U means univisited; will need to be looked at eventually.

class Expand(MRJob):
    
    # Note: change this to whatever source node actually is, I should change this to sys.arg variable eventually
    def mapper_init(self):
        self.source_node = "1"
        self.increment_counter('mapper', 'Total Mappers', 1)

    
    def mapper(self, _, line):
        
        # read the data for one node (comes as string; need to get relevant info)
        line = line.strip()
        node, data = line.split("\t")
        node_info = re.split("{.*}", data)[1]
        node_status = re.findall("[QVU]", node_info)[0]
        node_distance = re.findall("\d+", node_info)[0]
        node_distance = int(node_distance)
        frontier = []
        
        node = re.findall("\d", node)[0]
        
        # if node is in frontier we need to expand it
        if node_status == "Q":
            self.increment_counter('Node Debugging', 'Q Node', 1)
            
            # prepare edge data from current node 
            edges = re.findall('{.*}', data)[0]
            edges = edges.replace("'", "\"")
            edges = json.loads(edges)
            
            # change the status of current node to "visited"
            if node == self.source_node:
                yield node, (edges, 0, "V")
            else:
                yield node, (edges, node_distance, "V")
                
            # now set the edges of current node to "Q" for expansion
            for expanded_node in edges.keys():
                #print(type(expanded_node), expanded_node)
                expanded_node = str(expanded_node)
                yield expanded_node, ({}, node_distance+int(edges[expanded_node]), "Q")
                
        # if node is not in frontier then simply pass its unchanged values to the reducer
        elif node_status == "U" or node_status == "V":
            edges = re.findall('{.*}', data)[0]
            edges = edges.replace("'", "\"")
            edges = json.loads(edges)
            yield node, (edges, node_distance, node_status)
            self.increment_counter('Node Debugging', 'U or V Node', 1)
            
        # just for debugging purposes right now
        else:
            self.increment_counter('Node Debugging', 'wtf is this status?', 1)
            

    def reducer_init(self):
        self.source_node = "1"
        self.increment_counter('reducer', 'Total Reducers', 1)
            
    def reducer(self, keys, values):

        
        for key in keys:
            self.increment_counter('reducer', 'Found a key', 1)
            
            source_check = False
            node_distances = []
            node_U = False
            node_V = False
            node_Q = False

            # parse the lines of input
            for value in values:
                if key == self.source_node and value[0] != {}:
                    source_check = True
                    yield key, value
                    break

                if value[2] == "U":
                    node_edge_data = value[0]
                    known_distance = value[1]
                    status_update = "U"
                    node_U = True

                elif value[2] == "V":
                    node_edge_data = value[0]
                    known_distance = value[1]
                    status_update = "V"
                    node_V = True

                elif value[2] == "Q":
                    node_Q = True
                    self.increment_counter('reducer', 'Entered Q node in reducer', 1)
                    node_distances.append(value[1])
                
            # special case where source node is being re-evaluated   
            if node_Q == True and node_U == False and node_V == False:
                self.increment_counter('reducer', 'QT, UF, VF', 1)
                known_distance = min(node_distances)
                node_edge_data = {}
                status_update = "Q" 


            if node_Q == True and node_U == True:
                self.increment_counter('reducer', 'QT, UT, VF', 1)
                known_distance = min(node_distances)
                status_update = "Q" 

            elif node_Q == True and node_V == True:
                self.increment_counter('reducer', 'QT, UF, VT', 1)
                for distance in node_distances:
                    if distance < known_distance:
                        status_update = "Q" 
                        known_distance = distance

            if source_check == False:
                node_output = (node_edge_data, known_distance, status_update)
                yield key, node_output
    
    
if __name__ == '__main__':
    Expand.run()
````

````python
%%writefile ShortestPathInitialize.py
# ! /opt/anaconda/bin/python 
# -*- coding: utf-8 -*-

from __future__ import division
import re
import mrjob
import json
from mrjob.protocol import RawProtocol
from mrjob.job import MRJob
from mrjob.step import MRStep

# This is step 1 of 2 in implementing BFS in MRJob for parallelized graph traversal 

class Initialize(MRJob):
    
    def mapper_init(self):
        self.source_node = "1"
        self.increment_counter('mapper', 'total maps', 1)
    
    def mapper(self, _, line):
        # important for determining when to stop iterating over this MRJob
        self.increment_counter('mapper', 'total nodes', 1)
        line = line.strip()
        node_input, edges_input = line.split("\t")
        edges_input = edges_input.replace("'", "\"")
        edges_input = json.loads(edges_input)
        
        # make source node the first in the priority queue
        if node_input == self.source_node:
            yield node_input, (edges_input, 0, "Q")

        else:
            yield node_input, (edges_input, 0, "U")
        
    def reducer(self, key, values):
        self.increment_counter('reducer', 'total reducers', 1)
        for value in values:
            yield key, value
    
if __name__ == '__main__':
    Initialize.run()
````

<div style="margin:10px;border-left:5px solid #eee;">
<pre style="font-family:sans-serif;background-color:transparent">
Suppose you wanted to find the largest network distance from a single source,
i.e., a node that is the furthest (but still reachable) from a single source.

**How would you implement this task?**
**How is this different from finding the shortest path graph distances?**

This is essentially the opposite problem. When two paths to a node exist, you would prioritize the longest path instead of the shortest. The only question is whether you want the truly longest path or the longest path that doesn't cross the same node twice. Technically in any graph that is connected enough to be able to loop around itself, the longest path to a node will always be infinite distance away.

**Is this task more difficult to implement than the shortest path distance?**

The runtime of this task may be much longer as you may have to go through many iterations of MapReduce jobs to get to this longest node. Also, as stated above, you would have to be sure that your algorithm doesn't let a path pass the same node twice. Further, you still need to consider all path as a naive approach of always following the path with the greatest weight will not necessarily lead you to the longest path for the same reasons that it didn't lead us to the shortest path in our algorithm from the previous problems. The drawn out graph below illustrates both points.
